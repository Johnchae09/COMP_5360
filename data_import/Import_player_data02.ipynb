```python
import os
import glob
import pandas as pd

main_directory = os.path.normpath(os.getcwd() + os.sep + os.pardir)
data_directory = os.path.join(main_directory, 'data')
file_names = [f for f in os.listdir(data_directory) if os.path.isfile(os.path.join(data_directory, f))]
file_names
```

```python
files = []
for filename in file_names:
    if 'stats.csv' in filename:
        files.append(filename)

files = files[:-2]
files
```

```python
MVP_df = pd.read_csv(os.path.join(data_directory,'MVP player stats.csv'),header=1)
MVP_df
```

```python
season_mvp_dict = {}
for index,row in MVP_df.iterrows():
    season = row['Season']
    player = row['Player']
    season_short = season[2:]
    season_mvp_dict[season_short] = player

season_mvp_dict
```

```python
defensive_df = pd.read_csv(os.path.join(data_directory,'Defensive Player of the Year player stats.csv'),header=1)
defensive_df
```

```python
season_defensive_dict = {}
for index,row in defensive_df.iterrows():
    season = row['Season']
    player = row['Player']
    season_short = season[2:]
    season_defensive_dict[season_short] = player

season_defensive_dict
```

```python
pd.set_option('display.max_columns', None)
combine_df = pd.DataFrame()
for csv_file in files:
    df = pd.read_csv(os.path.join(data_directory,csv_file))
    df['Player'] = df['Player'].str.replace('*','')

    season = csv_file[:5]
    duplicate_players = df[df.duplicated('Player')]
    if not duplicate_players.empty:
        print("Duplicate players found:")
        print(duplicate_players['Player'])
    else:
        print("No duplicate players found.")
    if season in season_mvp_dict:
        mvp_player = season_mvp_dict[season]
        df['MVP'] = (df['Player'] == mvp_player).astype(int)
    if season in season_defensive_dict:
        defensive_player = season_defensive_dict[season]
        df['Defensive Player of the Year'] = (df['Player'] == defensive_player).astype(int)
    combine_df = pd.concat([combine_df,df])
    
mvp_count = (combine_df['MVP'] == 1).sum()
denfensive_count = (combine_df['Defensive Player of the Year'] == 1).sum()

print(mvp_count)
print(denfensive_count)
```

```python
combine_df.to_csv('combine_player_stat_w_MVP_defensive_player.csv',header=True,index=False)
```
