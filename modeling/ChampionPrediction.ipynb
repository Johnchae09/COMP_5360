```python
import os
import glob
import pandas as pd
import numpy as np
import seaborn as sns

main_directory = os.path.normpath(os.getcwd())
main_directory = os.path.dirname(main_directory)

data_directory = os.path.join(main_directory, 'team_data')

file = os.path.join(data_directory, 'normal_team_data.csv')
team_data = pd.read_csv(file)

file = os.path.join(data_directory, '2024_normal_data.csv')
team_2024_data = pd.read_csv(file)

team_2024_data.set_index(team_2024_data.columns[0], inplace=True) 
team_data.set_index(team_data.columns[0], inplace=True) 
team_data
```

```python

X = team_data.copy()
X = X.drop(columns=['W'])
#display(X)
y = team_data['W']

print(X.columns)
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns', 13)


sns.pairplot(team_data)
```

```python
from sklearn import tree, svm, metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.datasets import load_digits
from sklearn.preprocessing import scale

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')
```

```python
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.8)

# Create a LinearRegression object
model_test = LinearRegression()

# Fit the model to your data
model_test.fit(X_train, y_train)

y_pred = model_test.predict(X_test)

d = {'Prediction': y_pred,
     'Actual' : y_test}

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.8)

model = svm.SVC(kernel='rbf', C=1)

model.fit(X_train, y_train)   

y_pred = model.predict(X_test)

# Get indices of instances classified as 1
champ_i = np.where(y_pred == 16)[0]

# Select teams classified as 1 from the DataFrame
index = y_test.iloc[champ_i].index
champions = team_data.loc[index]
champions = champions[['W','W%','NRtg']]
display(champions)

print('Confusion Matrix:')
print(metrics.confusion_matrix(y_true = y_test, y_pred = y_pred))

print('Accuracy = ', metrics.accuracy_score(y_true = y_test, y_pred = y_pred))

print('Average difference = ', np.mean(np.abs(y_pred - y_test)) )
```

```python
C_values = np.linspace(1, 400, 100)
accuracies = {}
for C in C_values:
    modelC = svm.SVC(kernel='rbf', C=C)
    
    modelC.fit(X_train, y_train)
    
    y_pred = modelC.predict(X_test)
    #scores = cross_val_score(modelC, X, y, cv=5)
    
    #mean_accuracy = np.mean(scores)
    
    accuracies[C] = np.mean(np.abs(y_pred - y_test))

best_C = min(accuracies, key=accuracies.get)
best_accuracy = accuracies[best_C]

print("Best C: " , best_C)
print("Average Difference = ",best_accuracy)
```
