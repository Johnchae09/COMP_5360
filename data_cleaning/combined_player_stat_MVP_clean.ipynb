```python
import os
import glob
import pandas as pd
import numpy as np
main_directory = os.path.normpath(os.getcwd() + os.sep + os.pardir)
csv_directory = os.path.join(main_directory, 'data_import')

df = pd.read_csv(os.path.join(csv_directory,'combine_player_stat_w_MVP_defensive_player.csv'))
df
```

```python
name_map = {'Rk':'Rank','Pos':'Position','Tm':'Team','G':'Games','GS':'Games started','MP':'Minutes played per game'
             ,'FG':'Field goals per game','FGA':'Field goals attempt per game','FG%':'Field goal percent','3P':'3 point field goal per game',
             '3PA':'3 point field goal attempt per game', '3P%':'3 point field goal percentage','2P':'2 point field goal per game', '2PA':'2 point field goal attempt per game',
             '2P%':'2 point field goal percentage', 'eFG%':'Effective field goal percentage', 'FT':'Free throws per game',
             'FTA':'Free throw attempt per game','FT%':'Free throw percentage','ORB':'Offensive rebounds per game','DRB':'Defensive rebounds per game',
             'TRB':'Total rebounds per game','AST':'Assist per game','STL':'Steals per game','BLK':'Blocks per game','TOV':'Turn overs per game',
             'PF':'Personal fouls per game','PTS':'Points per game'}

renamed_df = df.rename(columns=name_map)
```

```python
renamed_df.columns
```

```python
result = renamed_df['Field goals per game'] / renamed_df['Field goals attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
renamed_df['Field goal percent'] = result

result = renamed_df['3 point field goal per game'] / renamed_df['3 point field goal attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
renamed_df['3 point field goal percentage'] = result

result = renamed_df['2 point field goal per game'] / renamed_df['2 point field goal attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
renamed_df['2 point field goal percentage'] = result

result = renamed_df['Free throws per game'] / renamed_df['Free throw attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
renamed_df['Free throw percentage'] = result

result = (renamed_df['Field goals per game'] + 0.5 *  renamed_df['3 point field goal per game'])/ renamed_df['Field goals attempt per game']
result = result.replace([np.inf, -np.inf, np.nan], 0)
renamed_df['Effective field goal percentage'] = result
```
